```python
import gym
import minerl
import gym_minecraft

env = gym.make('MinecraftCliffWalking1-v0')
env.init(start_minecraft=True)
```

# Random Actions:

```python
env.reset()

done = False
net_reward = 0
iter_count = 0
while not done:
    
    # Random actions
    action = env.action_space.sample()
        
    print(f'Action: {action}')
    obs, reward, done, info = env.step(action)

    net_reward += reward
    print("Total reward: ", net_reward)    

#     env.render() this is buggy?
    iter_count +=1
    if iter_count == 200:
        break
        
env.close()
```

```python
from baselines import deepq
```

```python
def callback(lcl, _glb):
    # stop training if reward exceeds 99
    is_solved = lcl['t'] > 100 and sum(lcl['episode_rewards'][-101:-1]) / 100 >= 99
    return is_solved


def train_dqn(env):
    act = deepq.learn(
        env,
        network='mlp',
        lr=1e-3,
        total_timesteps=100000,
        buffer_size=50000,
        exploration_fraction=0.1,
        exploration_final_eps=0.02,
        print_freq=10,
        callback=callback
    )
    print("Saving model to openai_baseline_basic_nav.pkl")
    act.save("openai_baseline_basic_nav.pkl")
```

```python
env.reset()
train_dqn(env)
```
